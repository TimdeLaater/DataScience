```python
penguins = sns.load_dataset("penguins")
penguins= penguins.dropna()
penguins.head()
```

```python
sns.pairplot(penguins)
```

Je ziet 16 clusters van 2 eigenschappen dus X 2 alleen ze staan er allemaal ook dubbel in dus 16X2/2 = 16

```python
features = ['bill_length_mm','bill_depth_mm','flipper_length_mm', 'body_mass_g']
km = KMeans(n_clusters=2, random_state=42).fit(penguins[features])
```

```python
penguins['cluster'] = km.predict(penguins[features])
```

```python
penguins.head()
```

```python
penguins.cluster.value_counts()
```

2 Clusters geeft de hoogste score hoe meer clusters hoe lager de score

```python
sns.pairplot(penguins, hue="cluster")
```

```python
sns.pairplot(penguins, hue="species")
```

```python
contingency_table = penguins.groupby(['species','cluster']).size().unstack('cluster', fill_value=0)
contingency_table
```
