```python
def calculate_accuracy(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    return (predictions == actuals).sum() / len(actuals)
```

```python
PremStats['Position'] = PremStats['Position'].replace(['MF,DF', 'MF,FW'],'MF')
PremStats['Position'] = PremStats['Position'].replace(['FW,MF', 'FW,DF'],'FW')
PremStats['Position'] = PremStats['Position'].replace(['DF,MF', 'DF,FW'],'DF')
PremStats.head()
```

```python
players_train, players_test = train_test_split(PremStats, test_size=0.3, stratify=PremStats['Position'], random_state=4)
print(players_train.shape, players_test.shape)
```

```python
features= ['Yellow_Cards']
dt_classification = DecisionTreeClassifier(max_depth = 3)
dt_classification.fit(players_train[features], players_train['Position'])
```

```python
predictionsOnTrainset = dt_classification.predict(players_train[features])
predictionsOnTestset = dt_classification.predict(players_test[features])

accuracyTrain = calculate_accuracy(predictionsOnTrainset, players_train.Position)
accuracyTest = calculate_accuracy(predictionsOnTestset, players_test.Position)

print("Accuracy on training set " + str(accuracyTrain))
print("Accuracy on test set " + str(accuracyTest))
```

```python

def plot_tree_classification(model, features, class_names):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          class_names=class_names,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
plot_tree_classification(dt_classification, features, PremStats.Position.unique())
```
