## Uitleg Data Science 

Data Science is het inzicht krijgen in ongestructeerde data door analyse algoritmes en systemen. Voor dit portofolio heb ik gewerkt in Jupyter Notebook met python als programmeer taal. Ik heb hiervoor de data van elke speler in de Engelse Premier League gebruikt. 

```python
import pandas as pd
import seaborn as sns
PremStats = pd.read_csv("../data/EPL_20_21.csv")
penguins = sns.load_dataset("penguins")
```

Hier heb ik vervolgends verschillende soorten analyses op los gelaten om een beter inzicht te krijgen in deze data. Voorbeelden hiervan zijn: Confidence interval, clustering, multivariate analysis, bivariate analysis and univariete analysis. Een belangrijk voorbeeld hiervan is de correlatie zoals hieronder. Dit laat per atribuut de correlatie met een ander atribuut wat een indicatie geeft hoeveel verband ze met elkaar hebben

```python
PremStats.corr().style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```
