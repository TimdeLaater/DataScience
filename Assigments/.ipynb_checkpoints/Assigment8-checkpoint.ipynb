```python
import pandas as pd
```

```python
PremStats = pd.read_csv("../data/EPL_20_21.csv")
```

```python
PremStats.head()
```

```python
import scipy.stats as st
confidence = 0.90
st.t.interval(confidence, len(PremStats)-1, loc=PremStats['Age'].mean(), scale=st.sem(PremStats['Age']))
```

```python

confidence = 0.95
st.t.interval(confidence, len(PremStats)-1, loc=PremStats['Age'].mean(), scale=st.sem(PremStats['Age']))
```

```python

confidence = 0.99
st.t.interval(confidence, len(PremStats)-1, loc=PremStats['Age'].mean(), scale=st.sem(PremStats['Age']))
```

```python

confidence = 0.99999
st.t.interval(confidence, len(PremStats)-1, loc=PremStats['Age'].mean(), scale=st.sem(PremStats['Age']))
```
