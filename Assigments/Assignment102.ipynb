```python
import pandas as pd
import seaborn as sns
PremStats = pd.read_csv("../data/EPL_20_21.csv")
penguins = sns.load_dataset("penguins")
```

De taak van een datascientist is om de data uit de dataset weer te geven zodat er verschillende conclusies uit kunnen getrokken worden en iets nuttig over de data gezegd kan worden. Om bijvoorbeeld te kijken of ervaring belangrijker is op een positie. Om dit voor elkaar te krijgen moeten eerst de posities goed samengevoegd worden omdat deze ook meerdere posities kan bevatten.

```python
PremStats.head()
```

```python
PremStats['Position'] = PremStats['Position'].replace(['MF,DF', 'MF,FW'],'MF')
PremStats['Position'] = PremStats['Position'].replace(['FW,MF', 'FW,DF'],'FW')
PremStats['Position'] = PremStats['Position'].replace(['DF,MF', 'DF,FW'],'DF')
PremStats.head()
```

Nu de posities er goed in staan moet hiervan de gemmidelde leeftijd berekend worden en naast elkaar gezet worden.

```python
sns.barplot(y="Age", x="Position", data=PremStats)
```

Als we naar de grafiek kijken zien we wel degelijk een verschil tussen de posities en kunnen we hier vervolgens conclusies uit op maken.

Er is te zien dat keepers gemiddeld veel ouder zijn wat kan beteken dat op de positie van keepers meer ervaring verlangt wordt dan ander posities. Ook is er een lichte trend te zien dat hoe veder je van eigen goal staat hoe jonger je bent..
